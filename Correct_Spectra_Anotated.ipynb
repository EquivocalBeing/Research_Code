```python
# The purpose of this code is to normalize spectra in order to acquire their element abundance.
# The code will first read in the data we give it, then normalize the spectra. 
# It will the run it through a different program called SPACE. This program is how the elemental data is acquired
# It will then compile all the elemental data into a file.


###############################################    Section 0     ####################################################################################################################################  

# This is just the various package need to run the code

from astropy.io import fits
import numpy as np
from matplotlib import pyplot as plt
from astropy.visualization import quantity_support
quantity_support()  # for getting units on the axes below  
from astropy import units as u #units
import astropy.wcs as fitswcs #wcs
from specutils import Spectrum1D, SpectralRegion #spectrum1D (specutils)
from specutils.fitting import fit_generic_continuum
import warnings
from astropy.modeling.polynomial import Polynomial1D
from astropy.modeling.fitting import LevMarLSQFitter
import pandas as pd
import subprocess


###############################################    Section 1     #####################################################################################################################################     

# This is where the data is declared. The variables i and j is also declare so that it can be 
# used in loops later in the code 

i = 0
j = 0

stars = pd.read_csv("/home/smu/Carlos/APOGEE_STDS/stdsa.dat", delimiter=',')

spec_name = stars['Spectra_name'] # These are all the names of the stars
spec_vel = stars['RV'] # This is the radial velocities of the said stars


# This loops through all the stars and operates on the individually
for i in range(0, len(spec_name)): 
    warnings.simplefilter('ignore')
    spec_path = "/home/smu/Carlos/APOGEE_STDS/" + spec_name[i] 
    file = fits.open(spec_path)
    specdata = file[0].data
    header = file[0].header
 
    # This makes it so python is able to properly read the data from the fits files
    wcs_data = fitswcs.WCS(header={'CDELT1': header['CDELT1'], 'CRVAL1': header['CRVAL1'],
                               'CUNIT1': header['CUNIT1'], 'CTYPE1': 'WAVE',
                               'CRPIX1': header['CRPIX1']})

    #set flux units
    flux= specdata * u.Jy
                                                # These both will be used to normalize the spectra
    spec = Spectrum1D(flux=flux, wcs=wcs_data)

    l0 = 8.498018E-7 # m
    cc = 299792.458 # km/second
    
    vel = spec_vel[i]  # km/s   # This grabs the specific RV from the star being ran

    lambda_shift = ((vel/cc * l0)) # units of meters


    lambda_shift_A = ((vel/cc * l0))*10000000000 # units of Angstroms


    x = (spec.spectral_axis - lambda_shift*u.m)
                                                        # These are used to shift the spectra according to its RV
    xA = (spec.spectral_axis - lambda_shift*u.m)*10000000000.0 
    y = spec.flux

    # This pice of the code is how the spectra get normalized
    spectrum = Spectrum1D(flux=y, spectral_axis=x)

    
    with warnings.catch_warnings():
        g1_fit = fit_generic_continuum(spectrum, median_window=3, model=Polynomial1D(3), fitter=LevMarLSQFitter(), 
                                   exclude_regions=[SpectralRegion(0.7 * u.um, 0.845 * u.um), SpectralRegion(0.8493 * u.um, 0.8503 * u.um),
                                                    SpectralRegion(0.8535 * u.um, 0.855 * u.um),SpectralRegion(0.866 * u.um, 0.867 * u.um),SpectralRegion(0.8813 * u.um, 0.885 * u.um)], weights=None)

        
 ###############################################    Section 2     ####################################################################################################################################     
    
    # This plots the pre-normalized data from the fits file
    fig = plt.figure(figsize = (7,5))
    plt.plot(x,y)    
    plt.plot(x,g1_fit(x), color = 'orange')
    plt.title(spec_name[i])
    plt.xlabel('m', fontsize = 14, fontweight = "bold")
    plt.ylabel('jy', fontsize = 14, fontweight = "bold")
    plt.xlim([8.44E-7,8.855E-7])
    plt.show()
    #plt.savefig('raw_spec.jpg')
    
    
    # This plots the normalized spectra
    spec_normalized = y / g1_fit(x)
    
    fig = plt.figure(figsize = (7,5))
    plt.plot(x,spec_normalized)
    plt.title(spec_name[i])
    plt.xlabel('m', fontsize = 14, fontweight = "bold")
    plt.ylabel('jy', fontsize = 14, fontweight = "bold")
    plt.xlim([8.44E-7,8.855E-7])
    plt.show()
    #plt.savefig('norm_spec.jpg')
    
###############################################    Section 3     ######################################################################################################################################     

    # This prints the data from the normalized spectra into a .out file
    with open ('/home/smu/Carlos/APOGEE_STDS/Nazgul/' + spec_name[i] + 'corrected.out','w') as f:
        for j in range(0, len(xA)):
            f.write(' ' + str(xA[j].value) + ' ' + str(spec_normalized[j]) + '\n')
    
  
    # This writes the information needed to run SPACE into a .par file
    with open ('/home/smu/Carlos/SPACE/par_files/' + spec_name[i] + '.par','w') as ff:
            
            # This tells SPACE where the normalize spectra data is via its path
            ff.write("obs_sp_file '/home/smu/Carlos/APOGEE_STDS/Nazgul/" + spec_name[i] + "corrected.out'\n")
            ff.write("#\n")
            
            # SPACE uses various libraries to find elemental abundances
            ff.write("GCOGlib '/home/smu/Carlos/SPACE/GCOG_library_v1.0/'\n")
            ff.write("#\n")
            ff.write("fwhm 0.8\n") # This is the full width halfmax
            ff.write("#\n")
            #ff.write("alpha\n") # This will combine and average all the metalicity data from all the found elements
            #                      that are denser than Fe into the element 'alpha' 
            ff.write("#\n")  
            # SPACE uses what is wavelength regime to tell it where and where not to look on the spectra
            #ff.write("wave_lims 8500 8855\n") # Full
            #ff.write("wave_lims 8505 8535 8547 8660 8670 8702\n") # Koch
            ff.write("wave_lims 8450 8493 8503 8535 8550 8660 8670 8800\n") # Campos
            #ff.write("wave_lims 8450 8490 8505 8535 8550 8655 8670 8750 8798 8813\n") # Delta
            #ff.write("wave_lims 8435 8493 8503 8535 8550 8655 8670 8800\n") # Epsilon Don't use, this is in testing
            ff.write("#\n")
            ff.write("#\n")
            ff.write("error_est\n")
            ff.write("#\n")
            ff.write("null_value 'NaN'\n")
            
    # This is python telling the terminal to execute SPACE       
    executable_path = "/home/smu/Carlos/SPACE/./SPACE-1.4a"
    
    arg2 = "/home/smu/Carlos/SPACE/par_files/" + spec_name[i] + ".par"   
    
    subprocess.run([executable_path, arg2])
    
###############################################    Section 5     ######################################################################################################################################      
    
    # This compiles the elemental data from all of the stars that were ran in the loop
    # I know it doesn't look pretty, but it work :/
    # The only with thing that will need to be changed are the paths if you want the file(s) in a different location
    # or if you put the data files in different locations
    '''
    awk_command = "awk 'NR==2 {print FILENAME} NR >1' /home/smu/Carlos/SPACE/par_files/" + spec_name[i] + "_TGM_ABD.dat > test"
    output = subprocess.check_output(awk_command, shell=True)
    print(output.decode())
    awk_command = "awk 'ORS=NR%2?FS:RS' < test > /home/smu/Carlos/SPACE/test1"
    output = subprocess.check_output(awk_command, shell=True)
    print(output.decode())
    awk_command = "awk '{gsub(/\/home\/smu\/Carlos\/SPACE\/par_files\//,\"\")}; 1' < /home/smu/Carlos/SPACE/test1 > /home/smu/Carlos/SPACE/test"
    output = subprocess.check_output(awk_command, shell=True)
    print(output.decode())
    awk_command = "awk '{gsub(/_TGM_ABD.dat/,\"\")}; 1' < /home/smu/Carlos/SPACE/test > /home/smu/Carlos/SPACE/apogee_out_files/" + spec_name[i] + ".out"
    output = subprocess.check_output(awk_command, shell=True)
    print(output.decode())

command = "cat /home/smu/Carlos/SPACE/apogee_out_files/*.fits.out > /home/smu/Carlos/SPACE/apogee_results/apogee_results_final_Delta_a.out"
output = subprocess.check_output(command, shell=True)
print(output.decode())
'''
```
