```python
import sys
import os, os.path
import numpy as np
import matplotlib.pyplot as plt
import csv
from scipy.stats import kde
import pandas as pd
import math

i = 0
myfile = pd.read_csv('/home/smu/Carlos/N6569/data.csv', comment='#', delimiter = ',')

clust_RA = 273.41166
clust_Dec = -31.82689

Name_SPACE = myfile['Spectra_name']
FeH_SPACE = myfile['Fe']
RV = myfile['best_rv']
Star_RA = myfile['ra']
Star_Dec = myfile['dec']
RV_close = myfile.loc[((np.sqrt(((myfile["ra"])-clust_RA)**2 + ((myfile["dec"])-clust_Dec)**2)) < 0.4),'best_rv']
FeH_close = myfile.loc[((np.sqrt(((myfile["ra"])-clust_RA)**2 + ((myfile["dec"])-clust_Dec)**2)) < 0.4),'Fe']


fig = plt.figure(figsize = (8,6))
plt.scatter(FeH_SPACE, RV, color = 'grey')
plt.scatter(FeH_close,RV_close, color = 'purple')
#plt.scatter(RV_box, color = 'orange')
plt.xlim(-2,0.5)
plt.ylim(-200,200)
plt.axhline(-40, linestyle = '-', color = 'gold')
plt.axhline(-60, linestyle = '-', color = 'gold')
plt.axvline(-1.05, linestyle = '-', color = 'gold')
plt.axvline(-0.7, linestyle = '-', color = 'gold')
plt.title("N6569", fontsize = 14)
plt.xlabel('FeH', fontsize = 14, fontweight = "bold")
plt.ylabel('RV', fontsize = 14, fontweight = "bold")
j = 1

#plt.savefig('N6569_plot.jpg')
```

```python
fig = plt.figure(figsize = (8,6))
for i in range(0, len(Name_SPACE)):
    loc = math.sqrt(((Star_RA[i] - clust_RA) ** 2) + ((Star_Dec[i] - clust_Dec) ** 2))
    
    plt.scatter(FeH_SPACE[i], RV[i], color = 'grey')
    plt.xlim(-2,0.5)
    plt.ylim(-200,200)
    plt.axhline(-40, linestyle = '-', color = 'gold')
    plt.axhline(-60, linestyle = '-', color = 'gold')
    plt.axvline(-1.05, linestyle = '-', color = 'gold')
    plt.axvline(-0.7, linestyle = '-', color = 'gold')
    plt.title("N6569", fontsize = 14)
    plt.xlabel('FeH', fontsize = 14, fontweight = "bold")
    plt.ylabel('RV', fontsize = 14, fontweight = "bold")
    if (loc < 0.4) :
        plt.scatter(FeH_SPACE[i], RV[i], color = 'purple')
        if (-60 < RV[i] < -40) & (-1.05 < FeH_SPACE[i] < -0.7):
            plt.scatter(FeH_SPACE[i], RV[i], color = 'orange') 
plt.savefig('N6569_plot.jpg')
```

```python
fig = plt.figure(figsize = (8,6))
for i in range(0, len(Name_SPACE)):
    loc = math.sqrt( ( (Star_RA[i] - clust_RA) ** 2) + ( (Star_Dec[i] - clust_Dec) ** 2)  )

    arc = loc * 60
    
    plt.scatter(arc, RV[i], color = 'purple')
    #plt.xlim(0,20)
    #plt.ylim(-200,200)
    plt.axvline(24, linestyle='--', color = 'violet',label ='')
    plt.axvline(7, linestyle='-', color = 'magenta', label = 'title radius')
    plt.axhline(-40, linestyle = '-', color = 'gold')
    plt.axhline(-60, linestyle = '-', color = 'gold')
    plt.ylabel('Radial Vel', fontsize = 14, fontweight = "bold")
    plt.xlabel('Distance from cluster (arcmin)', fontsize = 14, fontweight = "bold")
    if arc > 24:
        plt.scatter(arc, RV[i], color = 'grey')
    if (-60 < RV[i] < -40) & (-1.05 < FeH_SPACE[i] < -0.7):
        plt.scatter(arc, RV[i], color = 'orange', s=60)
plt.savefig('N6569_dist.jpg')
```
